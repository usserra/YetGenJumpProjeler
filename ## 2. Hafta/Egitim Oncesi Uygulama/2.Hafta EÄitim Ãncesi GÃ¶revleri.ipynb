Eşit paylaşım yapıldıktan sonra kalan şekerler parçalanacak. Örneğin, eğer toplamda 91 şekerleri varsa , her biri 30 tane alıp kalan 1 tane şekeri parçalayacaklar.
Kaç tane şekeri parçalayacaklarını hesaplayan bir Python kodu yazınız.

````python
berkcanSeker = int(input(""))
emirhanSeker = int(input(""))
eslemSeker = int(input(""))
toplamSeker = berkcanSeker + emirhanSeker + eslemSeker
parcalanacakSeker = toplamSeker % 3
bolunmusSeker = toplamSeker // 3
print(f"Her birinizin alacagi seker sayisi {bolunmusSeker} olarak hesaplanmistir.")
print(f"Parcalanacak seker sayisi {parcalanacakSeker} olarak hesaplanmistir.")
````
